```python
import pandas as pd
```

```python
df=pd.read_csv('cleaned_data_spacex_h.csv')
```

```python
df.head()
```

```python
df = df[df['success'] != 'unknown']
```

```python
df.shape
```

### Model Building

```python
y=df['success']
X=df.drop(columns='success')
```

```python
X.head()
```

```python
y.head()
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X.head()
```

```python
X['flight_number'].unique()
```

```python
df.dtypes
```

```python
X.dtypes
```

```python
X = X.loc[:, ~X.columns.str.startswith('payloads_')]
```

```python
X=X.astype(int)
```

```python
X.dtypes
```

```python
y.unique()
```

```python
y = y.map({'False': 0, 'True': 1})
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state=43)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.linear_model import LogisticRegression

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)  # increase max_iter if convergence warning
model.fit(X_train, y_train)

```

```python
train_preds = model.predict(X_train)

# 🔍 Predict on Test set
test_preds = model.predict(X_test)

# --------------------------------------
# 📊 Evaluation Function (for both sets)
# --------------------------------------
def evaluate_model(y_true, y_pred, dataset_name):
    print(f"📍 Evaluation on {dataset_name} set:")
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred))
    print("-" * 50)

# 🔎 Evaluate
evaluate_model(y_train, train_preds, "Training")
evaluate_model(y_test, test_preds, "Testing")
```
